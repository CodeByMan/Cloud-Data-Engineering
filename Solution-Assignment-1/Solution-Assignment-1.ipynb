# Q3 Write a program in which use all the operators we can use in Python”.
- (a) Arithmetic
- (b) Comparison
- (c) Assignment
- (d) Logical
- (e) Bitwise
- (f) Membership
- (g) Identity

```python
# (a) Arithmetic Operators
a = 8
b = 3
print("a =", a)
print("b =", b)

print()

print("Arithmetic Operators:")
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a % b =", a % b)

print()

print("Comparison Operators:")
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)

print()

print("Assignment Operators:")
c = 5
print("c =", c)
c += 2
print("c after c += 2:", c)

print()

print("Logical Operators:")
x = True
y = False
print("x and y:", x and y)
print("x or y:", x or y)
print("not x:", not x)

print()

print("Bitwise Operators:")
p = 6
q = 3
print("p & q =", p & q)
print("p | q =", p | q)
print("p ^ q =", p ^ q)

print()

print("Membership Operators:")
my_list = [1, 2, 3]
print("2 in my_list:", 2 in my_list)
print("5 not in my_list:", 5 not in my_list)

print()

print("Identity Operators:")
list1 = [1, 2, 3]
list2 = list1
list3 = [1, 2, 3]
print("list1 is list2:", list1 is list2)
print("list1 is list3:", list1 is list3)
print("list1 == list3:", list1 == list3)
```

# Q4 Calculate Percentage of Marks
. Completes the following steps of small task:
- Mention Marks of English, Islamiat and Maths out of 100 in 3 different variables
- Mention Variable of Total Marks and assign 300 to it
- Calculate Percentage

```python
english_marks = 85
islamiat_marks = 90
maths_marks = 95

total_marks = 300

obtained_marks = english_marks + islamiat_marks + maths_marks

percentage = (obtained_marks / total_marks) * 100

print("Marks obtained:")
print("English:", english_marks)
print("Islamiat:", islamiat_marks)
print("Maths:", maths_marks)
print("Total Marks:", total_marks)
print("Obtained Marks:", obtained_marks)
print("Percentage:", percentage, "%")
```

# Q 10 Write a program to calculate the electricity bill (accept number of unit from user) according to the following criteria:
   - upto 100 units: no charge
   - Next 200 units: Rs 5 per unit
   - After 200 units: Rs 10 per unit
   # For example:
   - If input unit is 350, total bill amount is Rs. 3500
   - If input unit is 97, total bill amount is Rs. 0
   - If input unit is 150, total bill amount is Rs. 750

```python
units = int(input("Enter number of units: "))

if units <= 100:
    bill = 0
elif units <= 300:
    bill = (units - 100) * 5
else:
    bill = 200 * 5 + (units - 300) * 10

print("Total bill amount is Rs.", bill)
```

# Q14 16. A shop will give discount of 10% if the cost of purchased quantity is more than 1000. Ask user for quantity. Suppose, one unit will cost 100. Judge and print total cost for user.

```python
apple_price = 50
banana_price = 30
milk_price = 120
bread_price = 80
eggs_price = 10

print("Welcome to Fresh Mart")
print("Price List: Apple = 50, Banana = 30, Milk = 120, Bread = 80, Eggs = 10")

apple_qty = int(input("Enter number of apples: "))
banana_qty = int(input("Enter number of bananas: "))
milk_qty = int(input("Enter number of milk packets: "))
bread_qty = int(input("Enter number of bread packets: "))
eggs_qty = int(input("Enter number of eggs: "))

apple_cost = apple_qty * apple_price
banana_cost = banana_qty * banana_price
milk_cost = milk_qty * milk_price
bread_cost = bread_qty * bread_price
eggs_cost = eggs_qty * eggs_price

total_cost = apple_cost + banana_cost + milk_cost + bread_cost + eggs_cost

if total_cost >= 1000:
    discount = total_cost * 0.10
    final_cost = total_cost - discount
    print("You got a 10% discount of Rs", discount)
    print("Your final bill after discount is Rs", final_cost)
else:
    print("Your total bill is Rs", total_cost)
    print("No discount applied")
```

# Q 15 A school has following rules for grading system.

- c. 45 to 50 - D
- d. 50 to 60 - C

# Q18  Write a program to check if a year is leap year or not. If a year is divisible by 4 then it is leap year but if the year is century year like 2000, 1900, 2100 then it must be divisible by 400.

```python
year = int(input("Enter year: "))

if year % 400 == 0:
    print("Leap year")
elif year % 100 == 0:
    print("Not a leap year")
elif year % 4 == 0:
    print("Leap year")
else:
    print("Not a leap year")
```
